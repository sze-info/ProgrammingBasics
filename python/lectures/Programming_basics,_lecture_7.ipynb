#File handling

1. Open a file, create a file object (*open()*, arguments: file name and opening method, eg. r --> read)
2. Performing operations with the object representing the file (e.g. reading the entire content as text: *read()* method)
3. Closing a file (with *close()* method)

Interesting data elements of the file object:
- name: name of the opened file
- mode: opening mode
- closed: have we closed it yet?

```python
# downloading (source: https://archive.org/stream/TheLittlePrince-English/littleprince_djvu.txt)
!wget -q https://raw.githubusercontent.com/sze-info/ProgrammingBasics/master/python/lectures/data/prince24.txt
f = open('prince24.txt', 'r')
print('File name: ', f.name, ', opening mode: ', f.mode, ', content:\n\n', f.read(), sep='')
f.close()
if f.closed:
  print('\nThe file is closed. Further operations are not possible.')
```

It is more appropriate to use *with*, because the file is automatically closed even if some error interrupted the program's execution.

```python
with open('prince24.txt', 'r') as f:
  print('File name: ', f.name, ', opening mode: ', f.mode, ', content:\n\n', f.read(), sep='')
  # There is no need for an explicit closing, *with* does this automatically at the end of the indented block
if f.closed:
  print('\nThe file is closed. Further operations are not possible.')
```

It is also possible to read fragments of a given length from the file if the maximum number of characters is given as an argument to read(). If the end of the file is reached, an empty string is returned.

```python
with open('prince24.txt', 'r') as f:
  while True:
    fragment = f.read(20)
    if fragment != '':
      print(fragment)
    else:
      break
```

However, more often than not, we need to read entire lines from the file. This can be done with the *readline()* function (the line also contains the end-of-line character).

```python
with open('prince24.txt', 'r') as f:
  number = 1
  while True:
    line = f.readline()
    if line != '':
      print(f'{number:02} : {line}', end='')
      number += 1
    else:
      break
```

The same more concisely, without calling readline(), by iterating through the file:

```python
with open('prince24.txt', 'r') as f:
  number = 1
  for line in f:
      print(f'{number:02} :', line.rstrip('\n')) # printing after cutting off the end-of-line mark
      number += 1
```

*readlines()* loads all the lines of the file into a list.

```python
with open('prince24.txt', 'r') as f:
  lines = f.readlines()
  number = 1
  for line in lines:
      print(f'{number:02} : {line}', end='')
      number += 1
```

We can also display the contents of the file word by word using the *split()* method of the string. (By default, it wraps along all white characters, but it can be re-parameterized.)

```python
with open('prince24.txt', 'r') as f:
  for word in f.read().split():
    print(word)
```

Now that we can read and write, we can also copy.

```python
def copy(dest: str, src: str):
  """Copying a file

  Parameters
  ----------
  dest : str
    Filename of the copy
  src : str
    Filename of the original
  """

  with open(src, 'r') as input:
    with open(dest, 'w') as output:
      for line in input:
        output.write(line)

def compare(a: str, b: str) -> bool:
  """Comparing the content of two files

  Parameters
  ----------
  a : str
    Filename of the 1st file
  b : str
    Filename of the 2nd file

  Returns
  -------
  bool
    True if the two files are identical, False otherwise.
  """

  with open(a) as inputA:
    with open(b) as inputB:
      return inputA.read() == inputB.read()

copy('whiskey2.txt', 'whiskey.txt')
if compare('whiskey.txt', 'whiskey2.txt'):
  print('The files are identical.')
else:
  print('The files are different.')

!rm whiskey2.txt
```

## Simple data file processing

We often use text files to store matrix-type data. Each line is placed in a separate text line, the columns are e.g. separated by a tab character.

```python
def save(mtx, file):
  with open(file, 'w') as f:
    for line in mtx:
      f.write('\t'.join(line) + '\n')

def load(file):
  mtx = []
  with open(file, 'r') as f:
    for line in f:
      mtx.append(line.rstrip('\n').split('\t'))
  return mtx

m = [
     ['a1', 'a2', 'a3'],
     ['b1', 'b2', 'b3'],
     ['c1', 'c2', 'c3']
]

save(m, 'matrix.txt')
!cat matrix.txt
print(load('matrix.txt'))
!rm matrix.txt
```

## Using CSV files

A more sophisticated solution is to use CSV (Comma Separated Values) files. This is also a structured text file, but most spreadsheet programs know and support it. Unfortunately, there are application-dependent differences, e.g. the content of the file will be different for different language settings. In the case of Hungarian Excel, the cells are separated by semicolons, in English Excel by commas. Python has a separate [module](https://docs.python.org/3/library/csv.html) for handling CSV files.

spreadsheet.png

```python
!wget -q https://raw.githubusercontent.com/sze-info/ProgrammingBasics/master/python/lectures/data/spreadsheet.csv
import csv

with open('spreadsheet.csv', 'r') as f:
  csvReader = csv.reader(f, delimiter=',', quotechar='"')
  mtx = []
  for row in csvReader:
    mtx.append(row)
print(mtx)

!cat spreadsheet.csv
```

We can get a similar result, but shorter, if we create a list directly from the CSV reader.

```python
import csv

with open('spreadsheet.csv', 'r') as f:
  csvReader = csv.reader(f, delimiter=',', quotechar='"')
  mtx = list(csvReader)
print(mtx)

!cat spreadsheet.csv
```

The task should now be to process a table saved from Excel. Let's create a function that determines the total number of products (Item) sold in a given date interval (OrderDate)!

```python
# https://www.contextures.com/xlsampledata01.html#data
!wget -q https://raw.githubusercontent.com/sze-info/ProgrammingBasics/master/python/lectures/data/sales.csv
import csv
from typing import List, Tuple # for documenting complex type hinting cases

def load(filename : str) -> Tuple[List[str], List[List[str]]]:
  with open(filename, 'r', encoding="utf-8") as f: # character encoding, if you need it
    csvReader = csv.reader(f)
    header = next(csvReader) # reading the next line
    data = []
    for row in csvReader:
      data.append(row)
  return header, data

def filter(header : List[str], data : List[List[str]], dateFrom : str, dateTo : str, item : str) -> int:
  helperDict = {head:idx for (idx, head) in enumerate(header)} # So that the contents of header cells can be used to index rows instead of numbers
  total = 0;
  for row in data:
    date = row[helperDict["OrderDate"]]
    if date>=dateFrom and date<=dateTo and item==row[helperDict["Item"]]:
      total += int(row[helperDict["Units"]])
  return total

sheet = load('sales.csv')
print(filter(*sheet, "11/8/2020", "4/10/2021", "Pen"))

!rm sales.csv
```

#Data visualization using [Matplotlib](https://matplotlib.org/stable/index.html)

[Pyplot](https://matplotlib.org/stable/tutorials/introductory/usage.html): A Python interface to Matplotlib for creating plots. The same type plots can be created with Matlab, too. The same services are available in several ways, but here we present only one of them (OO style) and focus only on the basic, most frequently used options.

kép.png

- [Figure](https://matplotlib.org/stable/api/figure_api.html) (fig): the complete figure, which includes the plot(s) and all their components (several plots can be placed next to each other in a grid layout) in a figure).
- [Axes](https://matplotlib.org/stable/api/axes_api.html) (ax): an element connected to a Figure, which are the components that display on a plot (one of the [Artist](https://matplotlib.org/stable/api/artist_api.html)s, which represents a drawing area, can be considered a kind of canvas or a piece of canvas (a Figure can have several plots, thus several Axes). It usually contains two [Axis](https://matplotlib.org/stable/api/axis_api.html) (axis) objects per chart, with divisions on them, their labels, axis names, etc. The drawing operations corresponding to the various chart types are the methods of this object (e.g. plot())
- [subplots()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html): the Pyplot method that creates the Figure and one or more of its plot areas.
- [plot()](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.plot.html): drawing *y* values as a function of *x* values on a diagram connected by a line, and/or marks the data points with markers ([Marker](https://matplotlib.org/stable/gallery/lines_bars_and_markers/marker_reference.html)).

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots() # By default, we get one plot (ax) in the figure (fig)
ax.plot([1, 2, 3, 4], [1, 4, 2, 3]); # List of x and then y values to be displayed. By default, they are connected by lines.
# The axes are assigned automatically.
```

We can ask for subtitles
- for the whole plot ([set_title()](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_title.html)), e.g. "Functions"
- for the x axis ([set_xlabel()](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_xlabel.html)), e.g. "x"
- for the y axis ([set_ylabel()](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_ylabel.html)), e.g. "f(x)"

A legend is displayed by the [legend()](https://matplotlib.org/stable/tutorials/intermediate/legend_guide.html) method. It looks for a place that hides the least amount of data. Its content depends on the value of the *label* parameter specified in the plot() call.

Several curves, arcs can be drawn on a plot. Matplotlib then automatically uses different settings to make the curves distinguishable.

```python
import matplotlib.pyplot as plt
import math
fig, ax = plt.subplots()
ax.set_title("Functions") # Title of the plot
ax.set_xlabel("x")         # Label of "x" axis
ax.set_ylabel("f(x)")      # Label of "y" axis
PRECISION = 50             # Precision: we generate that many data points to get smooth curves
# range() works only with integers. That's why we use it indirectly to generate PRECISION pieces of xi values in the investigated interval [0, 2pi]
x = [(x/(PRECISION-1))*(2.*math.pi) for x in range(PRECISION)]
sin_y = [math.sin(xi) for xi in x] # Determines the value of sin() function for every xi values
cos_y = [math.cos(xi) for xi in x] # The same with cos()
ax.plot(x, sin_y, label="sin()")   # The value of 'label' will be presented in the legend
ax.plot(x, cos_y, label="cos()")
ax.legend();                       # Shows the legend
```

If the automatically selected colors for the curves are not suitable for some reason (e.g. because the different colored curves cannot be clearly distinguished when printed on a monochrome laser printer), then we can set custom colors, line styles, and thicknesses.

Even the location of the legend can be changed, if e.g. we want to achieve a uniform appearance for several of our figures.

- [Line Styles](https://matplotlib.org/3.5.1/api/_as_gen/matplotlib.lines.Line2D.html#matplotlib.lines.Line2D.set_linestyle) (linestyle)
- [Colors](https://matplotlib.org/3.5.1/tutorials/colors/colors.html) (color)
- [Thickness](https://matplotlib.org/stable/api/_as_gen/matplotlib.lines.Line2D.html#matplotlib.lines.Line2D.set_linewidth) (linewidth)
- [Legend position](https://matplotlib.org/3.5.1/api/_as_gen/matplotlib.pyplot.legend.html)

```python
import matplotlib.pyplot as plt
import math
fig, ax = plt.subplots()
ax.set_title("Functions")
ax.set_xlabel("x")
ax.set_ylabel("f(x)")
PRECISION = 50
x = [(x/(PRECISION-1))*(2*math.pi) for x in range(PRECISION)]
sin_y = [math.sin(xi) for xi in x]
cos_y = [math.cos(xi) for xi in x]
# Red color (given by color constant), 1.5 points line width, dashed line style
ax.plot(x, sin_y, label="sin()", color="red", linewidth=1.5, linestyle="--")
# Black color (given by RGB color component intensities), 2.5 points line width, dotted line
ax.plot(x, cos_y, label="cos()", color=(0, 0, 0), linewidth=2.5, linestyle="dotted")
ax.legend(loc="upper right"); # Legend shown in the upper right corner
```

The data points can be marked separately to emphasize that only these data are actually available, the values that can be read along the connecting lines are only assumptions. (There were so many of them in the previous figures that it wouldn't have made sense to mark them separately.) Properties of the [markers](https://matplotlib.org/3.5.1/api/markers_api.html) to be displayed in the figure on the *marker* parameter can be set via *marker*.

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.plot([1, 2, 3, 4], [1, 4, 2, 3], marker="D"); # 'D' for diamond
```

The connecting lines may not even be necessary (because they would, for example, create the false illusion that the examined function is continuous). In this case, it is sufficient to indicate only the markers. The easiest way to do this is with plot()'s *fmt* parameter, which is a formatted string and consists of the following optional components:

- Marker, e.g. '.' for the dot, 'o' for the circle, '*' for the star, etc.
- Line style, e.g. '-' produces a solid line, '--' a dashed line, and ':' a dotted line.
- Color, e.g. 'r' produces red, 'g' green, 'b' blue.

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.plot([1, 2, 3, 4], [1, 4, 2, 3], "ob");
```

If you only want to show markers on the plot, it is probably more expressive to use the [scatter()](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.scatter.html) plot type (scatter plot, bubble plot). Appearance can be customized:

- marker: type of marker (e.g. circle for 'o')
- s: the size of the marker (given in square points)
- facecolor: fill color, e.g. "red".
- edgecolor: outline color, e.g. "black"

The size of the figure can be set to the desired value with the *figsize* argument tuple passed to subplots() (width and height of the figure in inches, by default 6.4 and 4.8). Obviously, this only makes sense together with the resolution (*dpi*), which is 100 by default.

```python
import matplotlib.pyplot as plt
import random
fig, ax = plt.subplots(figsize=(10, 10)) # 10x10 inch, approx. 25.4x25.4 cm
N = 15 # Number of data points per marker type
x1 = [random.uniform(0, 1) for xi in range(N)]
y1 = [random.uniform(0, 1) for yi in range(N)]
x2 = [random.uniform(0, 1) for xi in range(N)]
y2 = [random.uniform(0, 1) for yi in range(N)]
# Circle marker, increased marker size, red filling, black outline
ax.scatter(x1, y1, marker="o", s=500, facecolor="red", edgecolor="black")
# Square marker, green color
ax.scatter(x2, y2, marker="s", color="green");
```

The formatting of the data shown on the axes can also be influenced in countless ways, a more detailed description of which can be found here:

- [Tick locators](https://matplotlib.org/stable/gallery/ticks/tick-locators.html), placing ticks along axes
- [Tick formatters](https://matplotlib.org/stable/gallery/ticks/tick-formatters.html), formatting tick labels

For the sake of simplicity, only the [values and number of markers](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_xticks.html) and [labels](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_xticklabels.html) are defined with a list. If no separate labels are specified, the values are printed along the axis.

```python
import matplotlib.pyplot as plt
import math
fig, ax = plt.subplots()
PRECISION = 50
x = [(x/(PRECISION-1))*(2.*math.pi) for x in range(PRECISION)]
sin_y = [math.sin(xi) for xi in x]
# There will be 10 ticks on the x-axis, i.e. we divide the range [0, 2pi] into 9 intervals
XTICKS = 10
ax.set_xticks([(xi/(XTICKS-1))*(2*math.pi) for xi in range(XTICKS)])
ax.set_xticklabels([f"{2.*xi/(XTICKS-1):.1f}pi" for xi in range(XTICKS)])
# There will be 5 ticks on the y-axis, i.e. we divide the range [-1, +1] into 4 intervals
YTICKS = 5
ax.set_yticks([yi/2 for yi in range(-YTICKS//2, YTICKS-YTICKS//2)]) # -2, 3
ax.set_xlabel("x (rad)")
ax.set_ylabel("sin(x)")
ax.plot(x, sin_y);
```

Sometimes two curves with separate y-axes must be displayed in one chart area. A typical example of this is when the power and torque of an engine are plotted as a function of engine speed. The measurement data are now available in a CSV data file uploaded to the network, which must first be downloaded to our own machine (wget). The structure of the tabular data is not suitable for immediate display (see the content of the file after printing with cat), it must first be transformed according to the needs of pyplot (e.g. the revolutions are in the first column, of which we create a list so that the values can be displayed along the x-axis).

Since the two curves will use the same x-axis, we use the [twinx()](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.twinx.html) method to create a new one, placed at the "top" of the previous one, whose y-axis is placed on the right side.

The return value of the plot() methods is now needed to assign labels to the curves placed in the different plot areas in the legend.

```python
!wget -q https://raw.githubusercontent.com/sze-info/ProgrammingBasics/master/python/lectures/data/engine.csv # downloading
!cat engine.csv                      # printing
import csv
import matplotlib.pyplot as plt

with open("engine.csv", "r") as f:
  csvReader = csv.reader(f)
  header = next(csvReader)
  data = list(csvReader)

rpm = [float(x[0]) for x in data]    # 1st column
torque = [float(x[1]) for x in data] # 2nd column
power = [float(x[2]) for x in data]  # 3rd column

fig, ax1 = plt.subplots()
ax1.set_xlabel("RPM")
ax1.set_ylabel("Nm")
l1, = ax1.plot(rpm, torque) # torque plot on the first chart area
ax2 = ax1.twinx()           # request for the second chart area with common x axis
ax2.set_ylabel("kW")
l2, = ax2.plot(rpm, power, "-r") # power plot on the second chart area
# common legend
ax2.legend([l1, l2], ["Torque (left side)", "Power (right side)"]);
```

Sometimes the measurement data are located in a very wide range, which on a linearly scaled diagram leads to one of the following:
- it is not possible to read an exact value,
- not all data can be displayed on the chart at the same time.

This can be somewhat helped by setting the y-axis to a logarithmic scale ([set_yscale('log')](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_yscale.html)). In the following example, the same function is represented on a diagram drawn first with a linear and second with a logarithmic y-axis.

We also created two charts (the *nrows* parameter of subplots() specifies the number of rows, *ncols* the number of columns) in one figure so that the two methods can be easily compared.

Readability is further improved by showing a grid the background ([grid()](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.grid.html)).

```python
import matplotlib.pyplot as plt
# two charts in one row of a common figure
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(16, 9))
N = 25
# The x, and consequently the y values are intentionally not 0-started,
# since it could not be represented on the logarithmic scalex = [x for x in range(1, N)]
x = [x for x in range(1, N)]
y = [x**2 for x in range(1, N)] 
ax1.grid(True) # turning on the grid in the background
ax1.plot(x, y)
ax2.grid(True)
ax2.set_yscale('log') # The y-axis has a logarithmic scale
ax2.plot(x, y);
```

Various texts placed on charts can be [formatted](https://matplotlib.org/3.5.1/tutorials/text/text_props.html). For example, the caption of the chart can be typed in 24-point font (*fontsize*), blue (*color*), written with a sans-serif character set (*family*) and italic (*style*).

Not only the familiar labels can be formatted in this way, but you can also place [text](https://matplotlib.org/3.5.1/api/_as_gen/matplotlib.axes.Axes.text.html) anywhere on the diagram, which can be formatted in the same way.

The text can be entered with [(La)TeX formatting](https://matplotlib.org/3.5.1/tutorials/text/mathtext.html), which allows the easy usage of Greek letters and equations, special symbols (e.g. fractions, subtraction, integration), indexing, etc.

There is also a way to [create labels for specific data points](https://matplotlib.org/3.5.1/gallery/text_labels_and_annotations/annotation_demo.html).

```python
import matplotlib.pyplot as plt, random, math

# The distance of a plane point from the origin
def distance(x, y):
  return math.sqrt(x*x + y*y);

fig, ax = plt.subplots(figsize=(12, 8))
N = 1000
x = [random.gauss(0, 1) for xi in range(N)]
y = [random.gauss(0, 1) for xi in range(N)]
max_dist = -1 # The longest distance found so far
for xi, yi in zip(x, y):
  d = distance(xi, yi)
  if d > max_dist:
    max_dist = d
    dist_x, dist_y = xi, yi # Location of the farthest point
ax.axis([-6, +6, -4, +4]) # The smallest and largest values that can be plotted along the axes
# Setting font size, font color, character family, style (vertical, italic).
ax.set_title("Normal (Gaussian) distribution", fontsize="24", color="blue", family="sans-serif", style="italic")
# The first two parameters are the position of one corner of the rectangular area
# containing the text given in the coordinate system of the data. For us, the 
# value x=6 means the right edge of the diagram, and y=4 the top, so the reference
# point is the upper right corner of the diagram. This point is the location of 
# the right (horizontalalignment="right") upper (verticalalignment="top") corner
# of the rectangle. The caption was created with (La)TeX formatting to display 
# the Greek letters. 
# r-string: raw format, does not interpret escape sequences
ax.text(6, 4, r'$\mu=0,\ \sigma=1$', fontsize="14", horizontalalignment="right", verticalalignment="top")
# Labeling a data point that is furthest from the origin. *xy* is the location of
# the point to be marked, *xytext* is the location of the label on the diagram,
# and *arrowprops* specifies the properties of the arrow between the two: 
# *facecolor* is its color, and *shrink* is the amount of shrinking (so that it 
# does not touch the text and the point shown).
ax.annotate("Farthest point", xy=(dist_x, dist_y), xytext=(-5, -3), arrowprops={"facecolor": "black", "shrink": 0.05})
ax.grid(True)
ax.scatter(x, y);
```

Sometimes we also want to express measurement [errors](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.errorbar.html#matplotlib.axes.Axes.errorbar), the uncertainty of the data. The amount of error along both the x and y axes can be marked separately for each data point. The drawing of the plot is now entrusted to the [errorbar()](https://matplotlib.org/3.5.0/api/_as_gen/matplotlib.pyplot.errorbar.html) method.

```python
import matplotlib.pyplot as plt, random
N = 10
x = [x for x in range(N)]
y = [xi+random.gauss(0, 1) for xi in x]
# Now we also generate the measure of errors
ey = [abs(0.05*yi) for yi in y]
fig, ax = plt.subplots()
# Draw a curve with measurement errors and inaccuracies
# yerr is the amount of errors to be marked on the y axis, 
# capsize is the size of the perpendicular (now horizontal) closing line
ax.errorbar(x, y, yerr=ey, capsize=5);
```

The [histogram](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.hist.html) does not simply display the data, but is also prepared for certain data processing (e.g. how many data are in a given interval?). The example ([original version](https://matplotlib.org/stable/gallery/statistics/histogram_features.html)) shows the distribution of people's IQ scores (randomly generated data).

```python
import matplotlib.pyplot as plt
import random

mu = 100  # mean of distribution
sigma = 15  # standard deviation of distribution
x = [random.gauss(mu, sigma) for x in range(400)]
bins = 10 # The number of columns (measurement data ranges).
fig, ax = plt.subplots()
ax.set_xlabel('Smarts')
ax.set_ylabel('Probability density')
ax.set_title(r'Histogram of IQ: $\mu=' + str(mu) +', \sigma=' + str(sigma) +'$')
# If density=True, the columns do not show the raw numbers, but the value of the
# probability density function, i.e. the area of the histogram is 1.
ax.hist(x, bins, density=True);
```

If we want to show the proportion of each component, we often use a [pie chart](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.pie.html). It is possible to highlight some components from the others (*explode*), draw shadows (*shadow*) and write the corresponding value on individual items with the formatting we like (*autopct*).

```python
import matplotlib.pyplot as plt, random

# The components to be represented, the market players
labels = ["BMW", "Renault", "Mazda", "Jaguar", "Tata", "Kia"]
# Values of the components, here is the market share
x = [random.randint(1, 100) for _ in range(len(labels))]
# We will look for the market leader, we will highlight it
highlighted = x.index(max(x))
highlight = [0.1 if x==highlighted else 0 for x in range(len(x))]
fig, ax = plt.subplots(figsize=(12, 8))
ax.set_title("Market share")
ax.pie(x, labels=labels, explode=highlight, shadow=True, autopct='%1.1f%%');
```

Displaying the value of two-variable functions on a plane is not easy, but a [heatmap](https://matplotlib.org/stable/gallery/images_contours_and_fields/image_annotated_heatmap.html) (where the value of the function is given by the color of an area or a single point of a figure) can help. The meaning of the applied colors can be read from a [bar](https://matplotlib.org/stable/api/colorbar_api.html).

The function [Rastrigin](https://en.wikipedia.org/wiki/Rastrigin_function) is a frequently used test function to check and measure the operation of optimization algorithms. It is a continuous function that can be generalized to any number of independent variables. We will use its two-variable version here. It will take its global minimum at the value 0 of the independent variables.

```python
import matplotlib.pyplot as plt, math

# n-variable Rastrigin fv.
def rastrigin(*x):
  A = 10
  n = len(x)
  sum = A*2
  for xi in x:
    sum += xi**2-A*math.cos(2*math.pi*xi)
  return sum

PRECISION = 100
RANGE = 5.12
data = [[rastrigin(RANGE-2*RANGE*x/(PRECISION-1), RANGE-2*RANGE*y/(PRECISION-1)) for y in range(PRECISION)] for x in range(PRECISION)]
fig, ax = plt.subplots(figsize=(10, 10))
# The indexes of the data given in the form of a matrix would appear along the axes,
# we will replace this with the help of the extent parameter. Now the diagram shows
# the values of the two independent variables.
pos = ax.imshow(data, extent=[-RANGE, RANGE, -RANGE, RANGE])
# The size of the color bar can be reduced with the shrink parameter.
fig.colorbar(pos, ax=ax, shrink=0.75);
```
